```python
def get_cropped_image_if_2_eyes(image_path):
    
    image_color = cv2.imread(image_path)
    
    image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)

    face_cascade = cv2.CascadeClassifier('D:/Users/Py-MASTER/py-master/DataScience/CelebrityFaceRecognition/model/opencv/haarcascades/haarcascade_frontalface_default.xml')
    eye_cascade = cv2.CascadeClassifier('D:/Users/Py-MASTER/py-master/DataScience/CelebrityFaceRecognition/model/opencv/haarcascades/haarcascade_eye.xml')
    faces = face_cascade.detectMultiScale(image_gray, 1.3, 5)
    
    cropped_faces = []
    for (x, y, w, h) in faces:
        face_img = cv2.rectangle(image_color, (x,y), (x+w, y+h), (255,0,0), 2)
        roi_gray = image_gray[y:y+h, x:x+w]
        roi_color = face_img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)

        if len(eyes)>=2:
            cropped_faces.append(roi_color)
    return cropped_faces
```

### Wavelet Transformation

```python
import numpy as np
import pywt
import cv2    

def w2d(img, mode='haar', level=1):
    imArray = img
    #Datatype conversions
    #convert to grayscale
    imArray = cv2.cvtColor( imArray,cv2.COLOR_RGB2GRAY )
    #convert to float
    imArray =  np.float32(imArray)   
    imArray /= 255;
    # compute coefficients 
    coeffs=pywt.wavedec2(imArray, mode, level=level)

    #Process Coefficients
    coeffs_H=list(coeffs)  
    coeffs_H[0] *= 0;  

    # reconstruction
    imArray_H=pywt.waverec2(coeffs_H, mode);
    imArray_H *= 255;
    imArray_H =  np.uint8(imArray_H)

    return imArray_H
```

```python
results_dict = {'name': [], 'best params': [], 'best score': []}
```

```python
### using grid search


from sklearn.model_selection import GridSearchCV

params_lr = {
    'C': [0.1, 1, 10, 100, 1000],
    'solver': ['liblinear'], 
    'multi_class' : ['auto']
}

params_svc = {
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': [0.1, 1, 10, 100], 
    'C': [0.1, 1, 10, 100, 1000],
    'degree': [1,2,3,4,5,6]
}

params_rf = {
    'n_estimators': [int(x) for x in np.linspace(10,100,10)],
#     'max_features': ['auto', 'sqrt'],
#     'max_depth': [int(x) for x in np.linspace(1,10,2)],
#     'min_samples_leaf': [1, 4],
#     'min_samples_split': [2, 5],
#     'bootstrap': [True, False]
}
```

```python
from sklearn.model_selection import KFold, StratifiedKFold
kf = StratifiedKFold(n_splits=5)
def grid(estimator, params):
    gs = GridSearchCV(estimator=estimator, param_grid=params, cv=kf, verbose=2)
    gs.fit(X_train_scaled, y_train)
    results_dict['name'].append(str(estimator)[0: str(estimator).index('(')])
    results_dict['best params'].append(gs.best_estimator_)
    results_dict['best score'].append(gs.best_score_)
```

```python
grid(LogisticRegression(), params_lr)
```

```python
grid(SVC(probability=True), params_svc)
```

```python
grid(RandomForestClassifier(), params_rf)
```

```python

def predict_celebrity(path):
    name_class = []
    roi_color = get_cropped_image_if_2_eyes(path)
    if len(roi_color) != 0:
        for i in range(len(roi_color)):
            cropped_image = np.array(roi_color[i])
            im_har = w2d(cropped_image, 'db1', 5)
            scaled_wavelet = cv2.resize(im_har, (32,32))
            scaled_original = cv2.resize(roi_color[i], (32,32))
            combined = np.vstack((scaled_original.reshape(32*32*3, 1), scaled_wavelet.reshape(32*32, 1)))
            test = np.array(combined).reshape(1, len(combined)).astype(float)
            test_scaled = scaling(test.reshape(-1,1))
#             class_ = final_model.predict(test_scaled.reshape(1,-1))[0]
#             probability = final_model.predict_proba(test_scaled.reshape(1,-1))[0][class_]
            for name, class_ in celebrity_class_names.items():
                if class_ == final_model.predict(test_scaled.reshape(1,-1))[0]:
                    probability = final_model.predict_proba(test_scaled.reshape(1,-1))[0][class_]
                    name_class.append([name, probability])
#             name_class.append([class_, probability])
        return name_class
    return "Can't Detect eyes and face Properly"
```
